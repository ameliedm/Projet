## **Contexte historique**

Les marches aléatoires, en général, trouvent leur origine dans les travaux de Karl Pearson en 1905, qui les introduisit dans un contexte de diffusion des particules. Ce domaine a été enrichi par les contributions d’Andréi Markov avec sa théorie des chaînes de Markov.

La marche aléatoire minimale a été introduite en 2014 par Harbola, Kumar et Lindenberg. Elle constitue une variante de la célèbre marche de l’éléphant proposée au début des années 2000 par Schütz et Trimper pour étudier l’influence de la mémoire à long terme sur le comportement d’un marcheur aléatoire.

Dans la marche aléatoire minimale, le marcheur ne peut effectuer que deux types de mouvements : un pas en avant ou une pause, selon deux paramètres $p$ et $q$. Cette dynamique simple, mais dotée d’un mécanisme de sélection basé sur des positions passées, en fait un modèle intéressant pour analyser les effets de la mémoire dans les processus stochastiques.

Aujourd’hui, la marche aléatoire minimale est étudiée pour ses propriétés asymptotiques et ses applications en physique statistique, en biologie et en finance, où elle permet de modéliser des phénomènes dépendant d’un historique de décisions ou d’événements passés.

## **Introduction**  

La marche aléatoire minimale, très proche de la marche de l’éléphant, est un processus aléatoire à temps discret défini de la manière suivante :  

À l’instant 0, le marcheur est situé à l’origine, $ S_0 = 0 $. À l’instant 1, il se déplace vers la droite au point 1 avec une probabilité $ s $, ou reste au point 0 avec une probabilité $ 1 - s $, où $ 0 < s < 1 $.  

Ensuite, à l’instant $ n $, le marcheur choisit uniformément au hasard un instant $ k $ entre 1 et $ n $, puis :

$$
X_{n+1} =
\begin{cases} 
    \alpha_{n+1} & \text{si } X_k = 1 \\
    \beta_{n+1}  & \text{si } X_k = 0
\end{cases}
$$

où $ \alpha_{n+1} $ et $ \beta_{n+1} $ sont deux variables aléatoires indépendantes suivant des lois de Bernoulli $ B(p) $ et $ B(q) $ avec $ 0 < p, q < 1 $.  

En d’autres termes, nous avons :

$$
X_{n+1} = \alpha_{n+1} X_{U_n} + \beta_{n+1} (1 - X_{U_n})
$$

où $ U_n $ est une variable aléatoire discrète suivant une loi uniforme sur $ \{1, ..., n\} $ et $ \alpha_{n+1}, \beta_{n+1} $ et $ U_n $ sont mutuellement indépendantes.  

Le paramètre fondamental de la marche aléatoire minimale est défini par :  

$$
a = p - q
$$

La position du marcheur est alors donnée par la relation de récurrence suivante :

$$
S_{n+1} = S_n + X_{n+1}
$$

## Question 1

Montrer que l’on a :

$$ E[X_{n+1} | \mathcal{F}_n] = q + a \frac{S_n}{n} \quad \text{p.s.} $$
***

Tout d'abord, on sait que 
$$
X_{n+1} = \alpha_{n+1} X_{U_n} + \beta_{n+1} (1 - X_{U_n})
$$

donc on a 
$$
E[X_{n+1} | \mathcal{F}_n] = E[\alpha_{n+1} X_{U_n} + \beta_{n+1} (1 - X_{U_n})| \mathcal{F}_n]
$$

Les variables aléatoires $\alpha_{n+1}$ et $\beta_{n+1}$ sont indépendantes des $X_n$ et $U_n \in [\![1,n]\!]$,
on a alors 
$$
E[X_{n+1} | \mathcal{F}_n] = E[\alpha_{n+1}]E[X_{U_n}| \mathcal{F}_n] + E[\beta_{n+1}]E[1 - X_{U_n}| \mathcal{F}_n]
$$
De plus, $\alpha_{n+1}$ et $\beta_{n+1}$ sont deux variables aléatoires indépendantes suivant des lois de Bernoulli $ B(p) $ et $ B(q) $. Donc $ E[\alpha_{n+1}] = p $ et $ E[\beta_{n+1}] = q $
$$
\begin{aligned}
E[X_{n+1} | \mathcal{F}_n] &= pE[X_{U_n}| \mathcal{F}_n] + qE[1 - X_{U_n}| \mathcal{F}_n] \\
&= pE[X_{U_n}| \mathcal{F}_n] + q - qE[X_{U_n}| \mathcal{F}_n] \\
&= q + (p - q)E[X_{U_n}| \mathcal{F}_n] \\
&= q + aE[X_{U_n}| \mathcal{F}_n]
\end{aligned}
$$
Enfin, on sait que $X_{U_n}$ est uniformément distribué parmi $X_1$, $X_2$, ...$X_n$ puisque $U_n$ suit une loi uniforme sur $[\![1,n]\!]$. En sachant $\mathcal{F}_n$, $E[X_k| \mathcal{F}_n] = X_k$ pour $k\leq n$, l'espérance $E[X_{U_n}| \mathcal{F}_n]$ vaut donc la moyenne des $X_n \quad \text{p.s.}$, c'est à dire $\bar{X}_n = \frac{1}{n} \sum_{k=1}^{n} X_k \quad \text{p.s.}$. De plus, on a $S_n = \sum_{k=1}^{n} X_k $.

Au final on a donc :
$$
\begin{aligned}
E[X_{n+1} | \mathcal{F}_n] &= q + aE[X_{U_n}| \mathcal{F}_n] \\
&= q + a\frac{1}{n} \sum_{k=1}^{n} X_k \quad \text{p.s.}\\
&= q + a \frac{S_n}{n} \quad \text{p.s.}
\end{aligned}
$$

## Question 2

En déduire l’égalité presque sûre :

$$ E[S_{n+1} | \mathcal{F}_n] = q + \gamma_n S_n \quad \text{avec} \quad \gamma_n = \frac{n + a}{n}. $$
***

Pour commencer, on sait que $S_{n+1} = S_n + X_{n+1}$ donc 
$$
\begin{aligned}
E[S_{n+1} | \mathcal{F}_n] &= E[S_n + X_{n+1}| \mathcal{F}_n] \\
&= E[S_n | \mathcal{F}_n] + E[ X_{n+1}| \mathcal{F}_n]
\end{aligned}
$$
Etant donné que $S_n$ est $\mathcal{F}_n-mesurable$, on a $E[S_n | \mathcal{F}_n] = S_n$. De plus, on a vu à la question 1) que $ E[X_{n+1} | \mathcal{F}_n] = q + a \frac{S_n}{n} $.

On a donc :
$$
\begin{aligned}
E[S_{n+1} | \mathcal{F}_n] &= S_n + q + a \frac{S_n}{n} \\
&= q + S_n(\frac{a}{n} + 1) \\
&= q + S_n(\frac{a+n}{n}) \\
\end{aligned}
$$
Pour finir, nous avons :
$$
E[S_{n+1} | \mathcal{F}_n] = q + \gamma_n S_n \quad \text{avec} \quad \gamma_n = \frac{n + a}{n}, \forall n\geq 1. 
$$

## Question 3

Soit $ (M_n) $ la suite définie par :

$$ M_n = a_n S_n - q A_n $$

avec $ a_1 = 1 $, $ A_1 = 1 $ et pour $ n \geq 2 $,

$$ a_n = \prod_{k=1}^{n-1} \gamma_k^{-1} = \frac{\Gamma(n) \Gamma(1 + a)}{\Gamma(n + a)}, \quad A_n = \sum_{k=1}^{n} a_k. $$

Montrer que $ (M_n) $ est une martingale de carré intégrable.
***

Pour commencer, montrons que $ (M_n) $ est intégrable avant de montrer que c'est une martingale.
$$
\begin{aligned}
E[|M_n|] &= E[|a_n S_n - q A_n|] \\
&\leq E[|a_n S_n| + |q A_n|]
\end{aligned}
$$
On a cette inégalité par inégalité triangulaire.

De plus, $q > 0$, et $S_n$ et $A_n$ sont des sommes de termes positifs et $a_n$ est une multiplication de nombre positif, donc tous les termes sont positifs:
$$
E[|a_n S_n| + |q A_n|] = E[a_n S_n + q A_n]
$$
On a ensuite :
$$
\begin{aligned}
E[a_n S_n + q A_n] &= E[a_n S_n] + E[q A_n] \\
&= a_nE[S_n] + q A_n \quad \text{car} \quad a_n, q \quad \text{et} \quad A_n \quad \text{sont des constantes}
\end{aligned}
$$
De plus, $S_n = \sum_{k=1}^{n} X_k $ et les $X_k$ valent $0$ ou $1$ donc $S_n \leq n$ donc $E[S_n] \leq E[n]$ et n est une constante donc $E[n]=n$.

Donc :
$$
E[a_n S_n + q A_n] \leq a_nn + q A_n
$$
En conclusion :
$$
E[|M_n|] \leq a_nn + q A_n < \infty
$$
Donc $(M_n)$ est intégrable.

Montrons maintenant que c'est une martingale.
$$
\begin{aligned}
E[M_{n+1} | \mathcal{F}_n] &= E[a_{n+1}S_{n+1} - qA_{n+1}| \mathcal{F}_n] \\
&= E[a_{n+1}S_{n+1} | \mathcal{F}_n] - E[qA_{n+1}| \mathcal{F}_n] \\
&= a_{n+1}E[S_{n+1} | \mathcal{F}_n] - qA_{n+1} \quad \text{car} \quad a_{n+1}, q \quad \text{et} \quad A_{n+1} \quad \text{sont des constantes}
\end{aligned}
$$
On a vu à la question 2) que $E[S_{n+1} | \mathcal{F}_n] = q + \gamma_n S_n \quad \text{avec} \quad \gamma_n = \frac{n + a}{n} $. 

Donc :
$$
\begin{aligned}
E[M_{n+1} | \mathcal{F}_n] &= a_{n+1}(q + \gamma_nS_n) - qA_{n+1} \\
&= a_{n+1}q + a_{n+1}\gamma_nS_n - qA_{n+1} \\
&= a_{n+1}\gamma_nS_n + q(a_{n+1} - A_{n+1})
\end{aligned}
$$
Or $$ a_n = \prod_{k=1}^{n-1} \gamma_k^{-1} \quad \text{donc} \quad a_{n+1}\gamma_n = a_n\gamma_n^{-1}\gamma_n = a_n$$
de plus,
$$ \quad A_n = \sum_{k=1}^{n} a_k \quad \text{donc} \quad a_{n+1} - A_{n+1} = a_{n+1} - (A_n + a_{n+1}) = - A_n $$
Donc :
$$
\begin{aligned}
E[M_{n+1} | \mathcal{F}_n] &= a_nS_n - qA_n \\
&= M_n
\end{aligned}
$$
Donc $(M_n)$ est bien une martingale.

Pour finir, vérifions que $(M_n)$ est de carré intégrable.
$$
\begin{aligned}
E[M_n^2] &= E[(a_nS_n - qA_n)^2] \\
&= E[(a_nS_n)^2 - 2a_nS_nqA_n + (qA_n)^2] \\
&= E[(a_nS_n)^2] - E[2a_nS_nqA_n] + E[(qA_n)^2] \\
&= a_n^2E[S_n^2] - 2a_nqA_nE[S_n] + (qA_n)^2 \quad \text{car} \quad a_n, q \quad \text{et} \quad A_n \quad \text{sont des constantes}\\
&\leq a_n^2E[S_n^2] + 2a_nqA_nE[S_n] + (qA_n)^2 \quad \text{car} \quad 2a_nqA_nE[S_n] \quad \text{est un terme positif, comme nous l'avons déjà vu}
\end{aligned}
$$
On sait que $S_n = \sum_{k=1}^{n} X_k $ et les $X_k$ valent $0$ ou $1$ donc $S_n \leq n$ donc comme vu précédemment $E[S_n] \leq n$, de même $S_n^2 \leq n^2$ donc $E[S_n^2] \leq n^2$

En conclusion : 
$$
E[M_n^2] \leq a_n^2n^2 + 2a_nqA_nEn + (qA_n)^2 <  \infty
$$
Donc nous avons montré que $ (M_n) $ est une martingale de carré intégrable.

## Question 4

Prouver que son processus croissant $ \langle M \rangle_n $ est donné par :

$$ \langle M \rangle_n = (s - q)(1 - 2q) + q(1 - q) \sum_{k=1}^{n} a_k^2 + a(1 - 2q) \xi_n - a^2 \zeta_n $$

avec :

$$ \xi_n = \sum_{k=1}^{n-1} a_{k+1}^2 \frac{S_k}{k}, \quad \zeta_n = \sum_{k=1}^{n-1} a_{k+1}^2 \left(\frac{S_k}{k}\right)^2. $$


On veut calculer pour tour $n\geq 1$

$$
\begin{aligned}
\langle M \rangle_n &= \sum_{k=1}^n\mathbb{E}\left[\left.(M_k-M_{k-1})^2\right|\mathcal{F}_{k-1}\right]\\
&=\sum_{k=0}^{n-1}\mathbb{E}\left[\left.(M_{k+1}-M_{k})^2\right|\mathcal{F}_k\right]
\end{aligned}
$$

Soit $n\geq 1$ fixé, on a

$$
\begin{aligned}
M_{n+1}-M_n &=(a_{n+1}S_{n+1}-qA_{n+1})-(a_nS_n+qA_n) \\
&=a_{n+1}S_{n+1}-\overbrace{a_n}^{\gamma_na_{n+1}}S_n-q(\overbrace{A_{n+1}-A_n}^{a_{n+1}}) \\
&=a_{n+1}S_{n+1}-\gamma_na_{n+1}S_n-qa_{n+1} \\
&=a_{n+1}(S_{n+1}-\gamma_nS_n-q) \\
&=a_{n+1}\left(S_n+X_{n+1}-\left(1+\frac{a}{n}\right)S_n-q\right) \\
&=a_{n+1}\left(S_n+X_{n+1}-S_n-\frac{a}{n}S_n-q\right) \\
&=a_{n+1}\left(X_{n+1}-\left(q+\frac{a}{n}S_n\right)\right) \\
&=a_{n+1}\left(X_{n+1}-\mathbb{E}[X_{n+1}|\mathcal{F}_n]\right) \\
\end{aligned}
$$

Donc

$$
\begin{aligned}
\mathbb{E}[(M_{n+1}-M_n)^2|\mathcal{F}_n]&=a_{n+1}^2\mathbb{E}\left[\left(X_{n+1}-\mathbb{E}[X_{n+1}|\mathcal{F}_n]\right)^2|\mathcal{F}_n\right]\\
&=a_{n+1}^2\left(\mathbb{E}[X_{n+1}^2|\mathcal{F}_n]-\mathbb{E}[X_{n+1}|\mathcal{F}_n]^2\right)
\end{aligned}
$$
où on a $\mathbb{E}[X_{n+1}|\mathcal{F}_n]=q+a\frac{S_n}{n}$ et comme $X_{n+1}$ est à valeurs dans $\{0,1\}$, $\mathbb{E}[X_{n+1}^2|\mathcal{F}_n]=\mathbb{E}[X_{n+1}|\mathcal{F}_n]$. Alors, pour tout $n\geq1$

$$
\begin{aligned}
\mathbb{E}[(M_{n+1}-M_n)^2|\mathcal{F}_n]&=a_{n+1}^2\left(\mathbb{E}[X_{n+1}|\mathcal{F}_n]-\mathbb{E}[X_{n+1}|\mathcal{F}_n]^2\right)\\
&=a_{n+1}^2\left(q+a\frac{S_n}{n}-\left(q+a\frac{S_n}{n}\right)^2\right)\\
&=a_{n+1}^2\left(q+a\frac{S_n}{n}-q^2-a^2\frac{S_n^2}{n^2}-2qa\frac{S_n}{n}\right)\\
&=a_{n+1}^2\left(q(1-q)+a(1-2q)\frac{S_n}{n}-a^2\frac{S_n^2}{n^2}\right)
\end{aligned}
$$

De plus, en prennant $M_0=0$, on a

$$
\begin{aligned}
\mathbb{E}[(M_1-M_0)^2|\mathcal{F}_0]&=\mathbb{E}[(X_1-q)^2]\\
&=\mathbb{E}[X_1^2+q^2-2qS_1]\\
&=s+q^2-2qs
\end{aligned}
$$

Ainsi, pour $n\geq1$,

$$
\begin{aligned}
\langle M \rangle_n &= \sum_{k=0}^{n-1}\mathbb{E}\left[\left.(M_{k+1}-M_{k})^2\right|\mathcal{F}_k\right]\\
&= s+q^2-2qs + q(1-q)\sum_{k=1}^{n-1} a_{k+1}^2+a(1-2q)\sum_{k=1}^{n-1}a_{k+1}^2\frac{S_k}{k}-a^2\sum_{k=1}^{n-1}a_{k+1}^2\frac{S_k^2}{k^2}
\end{aligned}
$$

On pose $\forall n\geq 1$

$$
\xi_n=\sum_{k=1}^{n-1} a_{k+1}^2\frac{S_k}{k}\hspace{0.5cm}\hspace{0.5cm}\zeta_n=\sum_{k=1}^{n-1}a_{k+1}^2\frac{S_k^2}{k^2}
$$

Comme $a_1=1$, on peut écrire

$$
\begin{aligned}
s+q^2-2qs&=s+q^2-2qs-q(1-q)+q(1-q)a_1^2\\
&=s+2q^2-2qs-q+q(1-q)a_1^2\\
&=(s-q)(1-2q)+q(1-q)a_1^2\\
\end{aligned}
$$

et donc pour tout $n\geq 1$,

$$
\begin{aligned}
\langle M\rangle_n &=(s-q)(1-2q)+q(1-q)a_1^2+q(1-q)\sum_{k=2}^{n} a_k^2+a(1-2q)\xi_n-a^2\zeta_n\\
&=(s-q)(1-2q)+q(1-q)\sum_{k=1}^{n} a_k^2+a(1-2q)\xi_n-a^2\zeta_n
\end{aligned}
$$

## Question 6

En déduire que l’on a la convergence presque sûre :

$$ \lim_{n \to \infty} \frac{S_n}{n} = \frac{q}{1 - a} \quad \text{p.s.} $$


On a $S_n\leq n\,\forall n\geq 1$, donc

$$
\begin{aligned}
\langle M\rangle_n  \leq |(s-q)(1-2q)|+\sum_{k=1}^n a_k^2+|a(1-2q)|\sum_{k=2}^n a_k^2 \\
\Longrightarrow \langle M\rangle_n = \mathcal{O}\left(\sum_{k=1}^na_k^2\right)
\end{aligned}
$$

Or, quand $n\rightarrow\infty$, on a $\Gamma(n+a)\approx n^a\Gamma(n)$ donc 

$$
a_n=\frac{\Gamma(n)\Gamma(1+a)}{\Gamma(n+a)}\approx n^{-a}\Gamma(1+a)
$$

et quand $n\rightarrow \infty$, pour $0<2a<1$ ($0<a<1/2$ par hypothèse), $\sum_{k=1}^n n^{-2a}\approx \frac{n^{1-2a}}{1-2a}$, donc pour $n$ suffissamment grand, 

$$
\begin{aligned}
\sum_{k=1}^na_k^2 &\approx (\Gamma(1+a))^2\sum_{k=1}^n k^{-2a} \\
&\approx \frac{(\Gamma(1+a))^2}{1-2a}n^{1-2a}
\end{aligned}
$$

et ainsi

$$
\begin{aligned}
\langle M\rangle_n &=\mathcal{O}(\sum_{k=1}^na_k^2) \\
&=\mathcal{O}(n^{1-2a})
\end{aligned}
$$

D'autre part, la martingale $M_n$ vérifie 

$$
\begin{aligned}
&M_n^2=\mathcal{O}\left(\langle M\rangle_n log\langle M\rangle_n\right)\\
&\Longrightarrow M_n^2 =\mathcal{O}\left(n^{1-2a}(1-2a)log(n)\right) =\mathcal{O}\left(n^{1-2a}log(n)\right)
\end{aligned}
$$

Alors

$$
\begin{aligned}
\frac{M_n^2}{n^{2-2a}}&=\mathcal{O}\left(\frac{n^{1-2a}logn}{n^{2-2a}}\right)\\
&=\mathcal{O}\left(\frac{logn}{n}\right)
\end{aligned}
$$

Comme $\frac{logn}{n}\xrightarrow{n\rightarrow\infty}0$, alors 

$$
\begin{aligned}
\left(\frac{M_n}{n^{1-a}}\right)^2=\frac{M_n^2}{n^{2-2a}}\xrightarrow{n\rightarrow\infty}0\hspace{0.2cm}p.s. \\
\Longrightarrow \frac{M_n}{n^{1-a}}\xrightarrow{n\rightarrow\infty}0\hspace{0.2cm}p.s.
\end{aligned}
$$

Ainsi,

$$
\begin{aligned}
\frac{M_n}{n^{1-a}}&=\frac{a_nS_n-qA_n}{n^{1-a}}\\
&=na_n\frac{S_n}{n}-q\frac{A_n}{n^{1-a}}\xrightarrow{n\rightarrow\infty}0\hspace{0.2cm}p.s. \\
&\Longrightarrow \lim_{n\rightarrow\infty}na_n\frac{S_n}{n}=\lim_{n\rightarrow\infty}q\frac{A_n}{n^{1-a}}\hspace{0.2cm}p.s.
\end{aligned}
$$

Or, on a vu $a_n\approx n^{-a}\Gamma(1+a)$ quand $n\rightarrow\infty$ donc $na_n\xrightarrow{n\rightarrow\infty}n^{1-a}\Gamma(1+a)$. De plus,

$$
\begin{aligned}
A_n&=\sum_{k=1}^{n}a_k\\
&\approx\sum_{k=1}^{n}\Gamma(a+1) k^{-a}\\
&\approx\Gamma(a+1)\frac{n^{1-a}}{1-a}
\end{aligned}
$$

donc

$$
\lim_{n\rightarrow\infty}\frac{A_n}{n^{1-a}}=\frac{\Gamma(a+1)}{1-a}
$$

et finalement

$$
\begin{aligned}
&\lim_{n\rightarrow\infty}na_n\frac{S_n}{n}=\lim_{n\rightarrow\infty}q\frac{A_n}{n^{1-a}}\hspace{0.2cm}p.s.\\
&\Longrightarrow \Gamma(1+a)\lim_{n\rightarrow\infty}\frac{S_n}{n}=q\frac{\Gamma(1+a)}{1-a}\hspace{0.2cm}p.s.\\
&\Longrightarrow \lim_{n\rightarrow\infty}\frac{S_n}{n}=\frac{q}{1-a}\hspace{0.2cm}p.s.
\end{aligned}
$$

$$

D'après le Lemme de Toeplitz : 

Soit $(a_n)$ une suite de nombres réels positifs tel que $\sum_{k=1}^\infty(a_k) = \infty$. Si $A_n = \sum_{k=1}^n(a_k)$, et soit une suite $\lim_{n\rightarrow\infty}x_n = x$, alors on a $\lim_{n\rightarrow\infty}\frac{\sum_{k=1}^n(a_kx_k)}{A_n}=x$.

Dans notre cas, nous avons donc $x_n = \frac{S_n}{n}$ avec $\lim_{n\rightarrow\infty}x_n = \lim_{n\rightarrow\infty}\frac{S_n}{n} = \frac{q}{1-a}\hspace{0.2cm}p.s.$. Nous avons $(a_n^2)$ une suite de nombres réels positifs, avec $\sum_{k=1}^\infty(a_k^2) = \infty$. Pour conserver la notation du Lemme nous noterons $A_n = \sum_{k=1}^n(a_k^2)$ dans cette question.

Nous savons que 
$$
(a_n)=\frac{\Gamma(n)\Gamma(1+a)}{\Gamma(n+a)} 
$$
En utilisant l'inégalité assymptotique de la fonction Gamma : $\Gamma(n+a)\approx n^a\Gamma(n)$
$$
\begin{aligned}
(a_n)&\approx \frac{\Gamma(n)\Gamma(1+a)}{n^a\Gamma(n)}\\
&\approx \frac{\Gamma(1+a)}{n^a}
\end{aligned}
$$
donc 
$$
(a_n)^2\approx (\frac{\Gamma(1+a)}{n^a})^2
$$
Ainsi : 
$$
\sum_{k=1}^n(a_k^2)\approx (\Gamma(1+a))^2\sum_{k=1}^nk^{-2a}
$$
Or $\sum_{k=1}^nk^{-2a} \approx \frac{n^{1-2a}}{1-2a}$ pour $a<\frac{1}{2}$ donc :
$$
\sum_{k=1}^n(a_k^2)\approx (\Gamma(1+a))^2\frac{n^{1-2a}}{1-2a}
$$
Commençons avec 
$$
\frac{1}{n^{1-2a}}\sum_{k=1}^{n-1}a_{k+1}^2\frac{S_k}{k}
$$
Par le Lemme de Toeplitz,
$$
\begin{aligned}
\lim_{n\rightarrow\infty}\frac{\sum_{k=1}^{n-1}a_{k+1}^2\frac{S_k}{k}}{\sum_{k=1}^{n-1}a_{k+1}^2}&\approx \frac{\sum_{k=1}^{n}a_{k+1}^2\frac{S_k}{k}}{\sum_{k=1}^{n}a_{k+1}^2}\\
&\approx \frac{\sum_{k=1}^{n}a_{k+1}^2\frac{S_k}{k}}{A_n}\\
\end{aligned}
$$
or
$$
\begin{aligned}
\lim_{n\rightarrow\infty}\frac{\sum_{k=1}^{n}a_{k+1}^2\frac{S_k}{k}}{A_n}&=\lim_{n\rightarrow\infty}\frac{S_n}{n}\\
&=\frac{q}{1-a}\hspace{0.2cm}p.s.
\end{aligned}
$$
Nous avons :
$$
\begin{aligned}
\lim_{n\rightarrow\infty}\frac{1}{n^{1-2a}}\sum_{k=1}^{n-1}a_{k+1}^2\frac{S_k}{k}&=\frac{A_n}{n^{1-2a}}\frac{\sum_{k=1}^{n}a_{k+1}^2\frac{S_k}{k}}{A_n}\\
&=\frac{A_n}{n^{1-2a}}\frac{q}{1-a} \quad\text{avec} \quad A_n = \sum_{k=1}^n(a_k^2)\approx (\Gamma(1+a))^2\frac{n^{1-2a}}{1-2a}\\
&=(\Gamma(1+a))^2\frac{n^{1-2a}}{1-2a}\frac{1}{n^{1-2a}}\frac{q}{1-a}\\
&=(\Gamma(1+a))^2\frac{q}{(1-2a)(1-a)}
\end{aligned}
$$
et de la même manière avec $\lim_{n\rightarrow\infty}(\frac{S_n}{n})^2=(\frac{q}{1-a})^2$
$$
\begin{aligned}
\lim_{n\rightarrow\infty}\frac{1}{n^{1-2a}}\sum_{k=1}^{n-1}a_{k+1}^2(\frac{S_k}{k})^2&=\frac{A_n}{n^{1-2a}}\frac{\sum_{k=1}^{n}a_{k+1}^2(\frac{S_k}{k})^2}{A_n}\\
&=\frac{A_n}{n^{1-2a}}(\frac{q}{1-a})^2 \quad\text{avec} \quad A_n = \sum_{k=1}^n(a_k^2)\approx (\Gamma(1+a))^2\frac{n^{1-2a}}{1-2a}\\
&=(\Gamma(1+a))^2\frac{n^{1-2a}}{1-2a}\frac{1}{n^{1-2a}}(\frac{q}{1-a})^2\\
&=(\Gamma(1+a))^2\frac{q^2}{(1-2a)(1-a)^2}
\end{aligned}
$$

Ainsi :
$$
\lim_{n\rightarrow\infty}\frac{1}{n^{1-2a}}a(1-2q) \sum_{k=1}^{n-1}\frac{a_{k+1}^2}{n^{1-2a}}\frac{S_k}{k}-a^2\sum_{k=1}^{n-1}\frac{a_{k+1}^2}{n^{1-2a}}\frac{S_k^2}{k^2} = a(1-2q)(\Gamma(1+a))^2\frac{q}{(1-2a)(1-a)}-a^2(\Gamma(1+a))^2\frac{q^2}{(1-2a)(1-a)^2}
$$
D'un autre côté :
$$
\begin{aligned}
\lim_{n\rightarrow\infty}\frac{(s-q)(1-2q)}{n^{1-2a}}+q(1-q) \sum_{k=1}^n \frac{a_k^2}{n^{1-2a}}&=\lim_{n\rightarrow\infty}\frac{(s-q)(1-2q)}{n^{1-2a}}+q(1-q) \frac{A_n}{n^{1-2a}} \quad\text{avec} \quad A_n = \sum_{k=1}^n(a_k^2)\approx (\Gamma(1+a))^2\frac{n^{1-2a}}{1-2a}\\
&=\lim_{n\rightarrow\infty}\frac{(s-q)(1-2q)}{n^{1-2a}}+q(1-q) (\Gamma(1+a))^2\frac{n^{1-2a}}{1-2a}\frac{1}{n^{1-2a}}\\
&=\lim_{n\rightarrow\infty}\frac{(s-q)(1-2q)}{n^{1-2a}}+q(1-q) (\Gamma(1+a))^2\frac{1}{1-2a}\\
&=q(1-q) (\Gamma(1+a))^2\frac{1}{1-2a} \quad \text{car} \quad a<\frac{1}{2}
\end{aligned}
$$
Les limites de tous les termes étant finies, on peut les additioner pour trouver la limite totale et on a :
$$
\begin{aligned}
\lim_{n\rightarrow\infty}\frac{\langle M\rangle_n}{n^{1-2a}} &= q(1-q) (\Gamma(1+a))^2\frac{1}{1-2a} + a(1-2q)(\Gamma(1+a))^2\frac{q}{(1-2a)(1-a)}-a^2(\Gamma(1+a))^2\frac{q^2}{(1-2a)(1-a)^2}\\
&=\frac{\Gamma(1+a))^2}{1-2a}(q(1-q)+\frac{aq(1-2q)}{1-a}-\frac{a^2q^2}{(1-a)^2})\\
&=\frac{\Gamma(1+a))^2}{1-2a}(\frac{q(1-q)(1-a)^2+aq(1-2q)(1-a)-a^2q^2}{(1-a)^2}) \\
&=\frac{\Gamma(1+a))^2}{1-2a}(\frac{q}{(1-a)^2}\left((1-a)(1-q-a+aq+a-2aq)-a^2q\right)) \\
&=\frac{\Gamma(1+a))^2}{1-2a}(\frac{q}{(1-a)^2}\left((1-a)(1-q-aq)-a^2q\right)) \\
&=\frac{\Gamma(1+a))^2}{1-2a}(\frac{q}{(1-a)^2}\left(1-q-aq-a+aq+a^2q-a^2q\right)) \\
&=\frac{\Gamma(1+a))^2}{1-2a}(\frac{q}{(1-a)^2}(1-q-a)) \\
&=\frac{\Gamma(1+a))^2}{1-2a}(\frac{q(1-p)}{(1-a)^2}) \quad \text{car} \quad a=p-q 
\end{aligned}
$$

Pour conclure, 
$$
\lim_{n\rightarrow\infty}\frac{\langle M\rangle_n}{n^{1-2a}}=\frac{\Gamma(1+a))^2}{1-2a}(\frac{q(1-p)}{(1-a)^2})
$$
Nous avons bien 
$$ \lim_{n \to \infty} \frac{\langle M \rangle_n}{n^{1 - 2p + 2q}} = \sigma^2 \ell $$

avec :

$$ \sigma^2 = \frac{q(1 - p)}{(1 - a)^2}, \quad \ell = \frac{(\Gamma(1 + a))^2}{1 - 2a}. $$



## Question 7

Établir également la normalité asymptotique :

$$ \sqrt{n} \left( \frac{S_n}{n} - \frac{q}{1 - a} \right) \xrightarrow{L} \mathcal{N} \left( 0, \frac{\sigma^2}{1 - 2a} \right). $$
***

Nous allons utiliser le théorème limite centrale pour les martingales. Nous avons prouvé que $(M_n)$ est une martingale de carré intégrable et nous avons $\langle M\rangle_n$ son processus croissant. On a prouvé qu'il existe une suite déterministe de nombres réels $(b_n)_{n\geq1}$ avec $b_n=n^{1-2a}\,\quad \forall n\geq1$, croissante vers l'infini car $a<\frac{1}{2}$, tel que

$$
\lim_{n\rightarrow\infty}\frac{\langle M\rangle_n}{b_n}=\sigma^2 l<\infty
$$

Nous devons ensuite montrer que la condition de Lindeberg est respectée, donc que $\forall\epsilon>0$,
$$
\lim_{n\rightarrow\infty}\frac{1}{b_n}\sum_{k=1}^{n}E[(\Delta M_k)^2\mathbb{1}_{\{|\Delta M_k| \geq \epsilon \sqrt{b_n}\}}|\mathcal{F}_{k-1}]=0
$$
avec $\Delta M_k=M_k-M_{k-1}$

Or $\lim_{n\rightarrow\infty}\Delta M_n=0$ en particulier, $\forall\epsilon>0$, $\exists k*>0$ tel que $\forall k\geq k*, |\Delta M_k| < \epsilon \sqrt{b_n}$ donc 
$$
\lim_{n\rightarrow\infty}\frac{1}{b_n}\sum_{k=1}^{n}E[(\Delta M_k)^2\mathbb{1}_{\{|\Delta M_k| \geq \epsilon \sqrt{b_n}\}}|\mathcal{F}_{k-1}]=\lim_{n\rightarrow\infty}\frac{1}{b_n}\sum_{k=1}^{k*}E[(\Delta M_k)^2\mathbb{1}_{\{|\Delta M_k| \geq \epsilon \sqrt{b_n}\}}|\mathcal{F}_{k-1}]
$$
car après $k*$, l'indicatrice est nulle. Alors $\sum_{k=1}^{k*}E[(\Delta M_k)^2\mathbb{1}_{\{|\Delta M_k| \geq \epsilon \sqrt{b_n}\}}|\mathcal{F}_{k-1}]$ est une somme fini donc $\sum_{k=1}^{k*}E[(\Delta M_k)^2\mathbb{1}_{\{|\Delta M_k| \geq \epsilon \sqrt{b_n}\}}|\mathcal{F}_{k-1}]<C$, $C$ est une constante et $\lim_{n\rightarrow\infty}b_n=\infty$ donc $\lim_{n\rightarrow\infty}\frac{1}{b_n}\sum_{k=1}^{n}E[(\Delta M_k)^2\mathbb{1}_{\{|\Delta M_k| \geq \epsilon \sqrt{b_n}\}}|\mathcal{F}_{k-1}]<\lim_{n\rightarrow\infty}\frac{C}{b_n}=0$

On peut donc appliquer le théorème central limite pour les martingales et on a

$$
\frac{M_n}{\sqrt{b_n}}\xrightarrow{\mathcal{L}}\mathcal{N}(0,\sigma^2l)
$$

Alors,

$$
    \frac{M_n}{\sqrt{b_n}}=\frac{a_nS_n-qAn}{n^{\frac{1}{2}-a}}=\sqrt{n}\left(\frac{a_nS_n-qA_n}{n^{1-a}}\right)=\sqrt{n}\left(\frac{a_n}{n^{-a}}\frac{S_n}{n}-\frac{qA_n}{n^{1-a}}\right)\xrightarrow{\mathcal{L}}\mathcal{N}(0,\sigma^2l)
$$

Si $(c_n)_n$ et $(d_n)_n$ sont deux suites déterministes réelles qui convergent vers deux nombres réels $c$ et $d$ réspectivement et si $(X_n)_n$ est une suite de variables aléatoires qui converge au sens de la loi vers une variable aléatoire $X$, $X_n\xrightarrow{\mathcal{L}}X$. Alors on a 

$$
c_nX_n+d_n \xrightarrow{\mathcal{L}} cX+d
$$

Ainsi, on avait déjà calculé

$$
\lim_{n\rightarrow\infty}\frac{a_n}{n^{-a}}=\Gamma(1+a) \quad \text{car}\quad(a_n)\approx \frac{\Gamma(1+a)}{n^a}
$$

$$
\lim_{n\rightarrow\infty}q\frac{A_n}{n^{1-a}}=\frac{q\Gamma(1+a)}{1-a} \quad\text{car} \quad A_n\approx \Gamma(1+a)\frac{n^{1-a}}{1-a}
$$

et les suites $\left(\frac{a_n}{n^{-a}}\right)_n$ et $\left(\frac{qA_n}{n^{1-a}}\right)_n$ sont déterministes, donc on a bien

$$
\sqrt{n}\left(\Gamma(1+a)\frac{S_n}{n}-\Gamma(1+a)\frac{q}{1-a}\right)=\Gamma(1+a)\sqrt{n}\left(\frac{S_n}{n}-\frac{q}{1-a}\right)\xrightarrow{\mathcal{L}}\mathcal{N}(0,\sigma^2l)
$$

Or, soit $r\in\mathbb{R}^*$ et $X$ une variable aléatoire tel que $Y=rX\sim\mathcal{N}(0,\tilde{\sigma}^2)$, on peut montrer que $X\sim\mathcal{N}(0,\frac{\tilde{\sigma}^2}{r^2})$. En effet, si $\varphi:\mathbb{R}\rightarrow\mathbb{R}$ est une fonction quelconque, on a 

$$
\mathbb{E}[\varphi(X)]=\mathbb{E}\left[\varphi\left(\frac{Y}{r}\right)\right]=\int_{-\infty}^{+\infty}\varphi\left(\frac{y}{r}\right)\frac{1}{\sqrt{2\pi\tilde{\sigma}^2}}e^{-\frac{y^2}{2\tilde{\sigma}^2}}dy=\int_{-\infty}^{+\infty}\varphi(x)\frac{1}{\sqrt{2\pi\tilde{\sigma}^2}}re^{-\frac{r^2x^2}{2\tilde{\sigma}^2}}dx=\int_{-\infty}^{+\infty}\varphi(x)\frac{1}{\sqrt{2\pi\frac{\tilde{\sigma}^2}{r^2}}}e^{-\frac{x^2}{2\frac{\tilde{\sigma}^2}{r^2}}}dx
$$

donc la fonction densité de $X$ est $\frac{1}{\sqrt{2\pi\frac{\tilde{\sigma}^2}{r^2}}}e^{-\frac{x^2}{2\frac{\tilde{\sigma}^2}{r^2}}}$, c´est à dire, la fonction densité d'une variable aléatoire de loi $\mathcal{N}(0,\frac{\tilde{\sigma}^2}{r^2})$.

Ainsi, finalement on a 

$$
\Gamma(1+a)\sqrt{n}\left(\frac{S_n}{n}-\frac{q}{1-a}\right)\xrightarrow{\mathcal{L}}\mathcal{N}(0,\sigma^2l)\Longleftrightarrow \sqrt{n}\left(\frac{S_n}{n}-\frac{q}{1-a}\right)\xrightarrow{\mathcal{L}}\mathcal{N}\left(0,\frac{\sigma^2l}{(\Gamma(1+a))^2}\right)
$$

et donc avec l'expression explicite de $\sigma^2$ et $l$, on a

$$
 \sqrt{n}\left(\frac{S_n}{n}-\frac{q}{1-a}\right)\xrightarrow{\mathcal{L}}\mathcal{N}\left(0,\frac{\sigma^2}{1-2a}\right)
$$


## Question 8

Créer un premier code Python permettant de visualiser la convergence presque sûre et la normalité asymptotique de la marche aléatoire minimale où les paramètres $ p $ et $ q $ sont affectés par l’utilisateur avec $ p - q < 1/2 $, en faisant varier les valeurs de $ p $ et $ q $.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm
```

```python
# Fonction pour simuler la marche aléatoire minimale
def marche_aleatoire_minimale(n, p, q, s):
    X = np.zeros(n+1)  # Liste des X_k
    S = np.zeros(n+1)  # Liste des S_k
    # Initialisation de X[1] suivant une loi de Bernoulli de paramètre s
    X[1] = np.random.binomial(1, s)
    S[1] = X[1]  # Initialisation correcte de S[1]
    for i in range(2, n+1):
        k = np.random.randint(1, i)  # Tirage uniforme de k entre 1 et i-1
        alpha = np.random.binomial(1, p)
        beta = np.random.binomial(1, q)
        X[i] = alpha * X[k] + beta * (1 - X[k])  # Mise à jour de X_k
        S[i] = S[i-1] + X[i]  # Mise à jour de S_k
    return X, S

# Paramètres utilisateur
s = float(input("Entrez la valeur de s (0 < s < 1) : "))
p = float(input("Entrez la valeur de p (0 < p < 1) : "))
q = float(input("Entrez la valeur de q (0 < q < 1, avec p - q < 1/2) : "))
n = 5000  # Nombre d'itérations

if p - q >= 0.5 or p <= 0 or q <= 0 or p >= 1 or q >= 1:
    raise ValueError("Les paramètres doivent respecter 0 < p, q < 1 et p - q < 1/2.")

a = p - q

# Simulation

X, S = marche_aleatoire_minimale(n, p, q, s)  # Récupérer les deux valeurs correctement

# Convergence presque sûre : S_n / n
plt.figure(figsize=(10, 5))
plt.plot(np.arange(1, n+2), S / np.arange(1, n+2), label=r'$S_n / n$', color='blue')  # Dimensions alignées
plt.axhline(y=q / (1 - a), color='red', linestyle='--', label=r'$q / (1-a)$')
plt.xlabel("n")
plt.ylabel(r"$S_n / n$")
plt.title("Convergence presque sûre de la marche aléatoire minimale")
plt.legend()
plt.grid()
plt.show()


# Nombre de simulations
num_simulations = 4000  
final_values = np.zeros(num_simulations)

# Effectuer plusieurs simulations
for i in range(num_simulations):
    X, S = marche_aleatoire_minimale(n, p, q, s)
    final_values[i] = S[-1] / n  # Stocker la valeur finale normalisée

# Affichage de l'histogramme
plt.figure(figsize=(10, 5))
plt.hist(final_values, bins=50, density=True, alpha=0.6, color='g', label="Densité empirique")

# Ajout de la densité théorique normale
mu = q / (1 - a)
sigma = np.sqrt((q * (1 - p)) / ((1 - 2 * a) * n))  # Approximation de la variance asymptotique
x = np.linspace(min(final_values), max(final_values), 100)
plt.plot(x, norm.pdf(x, loc=mu, scale=sigma), 'r-', lw=2, label="Densité normale théorique")
plt.xlabel(r"Valeurs de $\frac{S_n}{n}$")
plt.ylabel("Densité")
plt.title(r"Approximation de la loi limite de $\frac{S_n}{n}$")
plt.legend()
plt.grid()
plt.show()

# Calcul des valeurs centrées et normali0sées
normalized_values = np.sqrt(n) * (final_values - q / (1 - a))

# Affichage de l'histogramme
plt.figure(figsize=(10, 5))
plt.hist(normalized_values, bins=50, density=True, alpha=0.6, color='b', label="Densité empirique")

# Ajout de la densité théorique normale
sigma_norm = np.sqrt(q * (1 - p) / (1 - 2 * a))  # Variance théorique
x = np.linspace(min(normalized_values), max(normalized_values), 100)
plt.plot(x, norm.pdf(x, loc=0, scale=sigma_norm), 'r-', lw=2, label="Densité normale théorique")

plt.xlabel(r"Valeurs de $\sqrt{n} \left(\frac{S_n}{n} - \frac{q}{1-a}\right)$")
plt.ylabel("Densité")
plt.title("Approximation de la normalité asymptotique centrée")
plt.legend()
plt.grid()
plt.show()
```

1. **Simulation de la marche aléatoire minimale**

- La marche commence à $S_0 = 0$.
- À chaque instant $i$, on choisit aléatoirement un instant $k$ parmi $\{1, \dots, i-1\}$.

$$\sigma^2 = \frac{q(1 - p)}{(1 - a)^2} = \frac{(1 - p)(1 - p)}{(1 - (2p - 1))^2} = \frac{(1 - p)^2}{(2 - 2p)^2} = \frac{(1 - p)^2}{4(1 - p)^2} = \frac{1}{4}.$$  

Ainsi, nous obtenons la normalité asymptotique suivante :

$$\sqrt{n} \left( \frac{S_n}{n} - \frac{1}{2} \right) \xrightarrow{L} \mathcal{N} \left( 0, \frac{1}{4(1 - 2(2p - 1))} \right) = \mathcal{N} \left( 0, \frac{1}{4(2 - 4p)} \right).$$

## Question 10

On propose d’estimer le paramètre $ p $) de la marche aléatoire minimale par l’estimateur du quasi-maximum de vraisemblance défini, pour tout $ n \geq 2 $, par :

$$ \hat{p}_n = \frac{\sum_{k=1}^{n-1} (2\bar{X}_k - 1)(X_{k+1} + \bar{X}_k - 1)}{\sum_{k=1}^{n-1} (\bar{X}_k - 1)^2} $$

où la moyenne empirique $ \bar{X}_n = S_n / n $.  
Créer un second code Python permettant de visualiser la convergence presque sûre :

$$ \lim_{n \to \infty} \hat{p}_n = p \quad \text{p.s.} $$

ainsi que la normalité asymptotique :

$$ \sqrt{\log n} (\hat{p}_n - p) \xrightarrow{L} \mathcal{N} \left( 0, \frac{3 - 4p}{4} \right). $$


```python
def estimateur_p(S, X, n):
    X_moy = np.zeros(n-1)
    X_ici = X[2:n+1]
    for i in range(n-1):
        X_moy[i] = S[i] / (i+1)  # Moyenne empirique pour chaque k
    num = np.sum((2 * X_moy - 1) * (X_ici + X_moy - 1))
    denom = np.sum((2 * X_moy - 1) ** 2)
    return num / denom if denom != 0 else 0

# Exemple d'utilisation
n = 10000
p = 0.4
q = 0.6
s = 0.2

# Calcul de l'estimateur pour différentes valeurs de n
X, S = marche_aleatoire_minimale(n, p, q, s)
p_estim = np.array([estimateur_p(S, X, k) for k in range(2, n+1)])
a = 2 * p - 1
sigma2 = (3 - 4 * p) / 4  # Variance théorique
sigma = np.sqrt(sigma2)

# Convergence presque sûre de \hat{p}_n
plt.figure(figsize=(10, 5))
plt.plot(np.arange(2, n+1), p_estim, label=r'$\hat{p}_n$')
plt.axhline(y=p, color='red', linestyle='--', label=r'Valeur réelle $p$')
plt.xlabel("n")
plt.ylabel("\hat{p}_n")
plt.title("Convergence presque sûre de l'estimateur du quasi-maximum de vraisemblance")
plt.legend()
plt.grid()
plt.show()
```

```python
# Paramètres
n = 1000  # Nombre d'étapes dans la marche aléatoire
p = 0.4
q = 0.6
s = 0.2
n_simulations = 500  # Nombre de simulations

# Liste pour stocker les valeurs normalisées finales
stats_sim = []

for _ in range(n_simulations):
    # Simuler une marche aléatoire
    X, S = marche_aleatoire_minimale(n, p, q, s)
    
    # Calculer l'estimateur de p
    p_estim = np.array([estimateur_p(S, X, k) for k in range(2, n+1)])
    
    # Calculer la statistique normalisée à l'instant final (n)
    normalised_stat_final = np.sqrt(np.log(n)) * (p_estim[-1] - p)
    stats_sim.append(normalised_stat_final)

# Histogramme des valeurs obtenues
plt.figure(figsize=(10, 5))
plt.hist(stats_sim, bins=50, density=True, alpha=0.6, color='b', label='Densité empirique')

# Tracer la densité de la loi normale théorique
sigma2 = (3 - 4 * p) / 4  # Variance théorique
sigma = np.sqrt(sigma2)
x = np.linspace(min(stats_sim), max(stats_sim), 100)
plt.plot(x, stats.norm.pdf(x, 0, sigma), 'r-', lw=2, label='Loi normale théorique')

# Mise en forme du graphe
plt.xlabel(r"$\sqrt{\log(n)} (\hat{p}_n - p)$")
plt.ylabel("Densité")
plt.title("Normalité asymptotique de l'estimateur sur plusieurs simulations")
plt.legend()
plt.grid()
plt.show()
```
